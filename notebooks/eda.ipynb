```python
profit = pd.read_sql(
    "SELECT * FROM profitandloss",
    conn
)

profit.head()
```

```python
profit.isnull().sum()
```

```python
profit.describe()
```

```python
top_sales = profit.sort_values(
    by="sales",
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

plt.bar(
    top_sales["company_id"],
    top_sales["sales"]
)

plt.title("Top 10 Companies by Sales")
plt.xlabel("Company")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

```python
top_profit = profit.sort_values(
    by="net_profit",
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

plt.bar(
    top_profit["company_id"],
    top_profit["net_profit"]
)

plt.title("Top 10 Companies by Net Profit")
plt.xlabel("Company")
plt.ylabel("Net Profit")

plt.xticks(rotation=45)

plt.show()
```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

conn = sqlite3.connect("../db/nifty100.db")

print("Connected Successfully!")
```

```python
market = pd.read_sql(
    "SELECT * FROM market_cap",
    conn
)

market.head()
```

```python
market.info()
```

```python
market.describe()
```

```python
top_market = market.sort_values(
    by="market_cap_crore",
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

plt.bar(
    top_market["company_id"],
    top_market["market_cap_crore"]
)

plt.title("Top 10 Companies by Market Capitalization")
plt.xlabel("Company")
plt.ylabel("Market Cap (Crores)")
plt.xticks(rotation=45)

plt.show()
```

```python
top_pe = market.sort_values(
    by="pe_ratio",
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

plt.bar(
    top_pe["company_id"],
    top_pe["pe_ratio"]
)

plt.title("Top 10 Companies by PE Ratio")
plt.xlabel("Company")
plt.ylabel("PE Ratio")
plt.xticks(rotation=45)

plt.show()
```
